```python
%matplotlib notebook
%matplotlib nbagg
import numpy as np
#import matplotlib
#matplotlib.use('nbagg')
import matplotlib.pyplot as plt
from matplotlib import animation
import random

fig = plt.figure()
 
xlim = [0,100]
X, Y = [], []
 
def plot(data):
    plt.cla()                   # å‰ã®ã‚°ãƒ©ãƒ•ã‚’å‰Šé™¤
    
    Y.append(random.random())   # ãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ
    X.append(len(Y))
    
    if len(X) > 100:            # æç”»ç¯„å›²ã‚’æ›´æ–°
        xlim[0]+=1
        xlim[1]+=1
    
    plt.plot(X, Y)              # æ¬¡ã®ã‚°ãƒ©ãƒ•ã‚’ä½œæˆ
    plt.title("sample animation (real time)")
    plt.ylim(-1,2)
    plt.xlim(xlim[0],xlim[1])
    
 
# 10msã”ã¨ã«ploté–¢æ•°ã‚’å‘¼ã³å‡ºã—ã¦ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã‚’ä½œæˆ
ani = animation.FuncAnimation(fig,plot,frames = 132,interval=10, repeat=False)
#ani.save('sample2.gif', writer='imagemagick')
plt.show()
```
